```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
data = pd.read_csv('Glassdoor_Gender_Pay_Gap.csv')
data.info()
Job=data['JobTitle']
Gender=data['Gender']
Age=data['Age']
PerfEval=data['PerfEval']
Education=data['Education']
Dept=data['Dept']
Seniority=data['Seniority']
BasePay=data['BasePay']
Bonus=data['Bonus']
income=BasePay+Bonus
data.info()
```

```python
data.columns
```

```python
data.count()
```

```python
data['income'] = data['BasePay'] + data['Bonus']
```

## Correlation matrix of the variables

```python
plt.figure(figsize=(15,15))
sb.heatmap(data.corr(),annot = True,fmt = ".2f",cbar = True)
plt.xticks(rotation=90)
plt.yticks(rotation = 0)
```

## Plots for income

```python
f, axes = plt.subplots(3, figsize=(12, 12))
sb.boxplot(data = income, orient = "h", ax = axes[0])
sb.histplot(data = income, ax = axes[1])
sb.violinplot(data = income, orient = "h", ax = axes[2])
```
